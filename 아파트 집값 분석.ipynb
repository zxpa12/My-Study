# groupby 로 데이터 집계하기
  - groupby를 통해 데이터를 그룹화해서 연산을 해봅니다.

```python
df_last
```

```python
# 지역명으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 합니다
# df.groupby(["인덱스 컬럼명"]){"계산할 컬럼 값"}.연산()
df_last.groupby(["지역명"])["평당분양가격"].mean()
```

## melt로 Tidy data 만들기

```python
df_first.head(1)
```

```python
df_first_melt= df_first.melt(id_vars = "지역")
df_first_melt.head()
```

```python
df_first_melt= df_first.melt(id_vars = "지역", var_name = "기간", value_name = "평당분양가격")
df_first_melt.head()
```

```python
df_first_melt.columns = ["지역명", "기간", "평당분양가격"]
df_first_melt.head(1)
```

## 연도와 월을 분리하기

```python
date = "2013년12월"
date
```

```python
# split을 통해 "년"을 기준으로 텍스트를 분리
date.split('년')
```

```python
# 리스트의 인덱싱을 사용해서 연도만 가져옴
date.split("년")[0]
```

```python
# replace를 사용해서 월을 제거
date.split("년")[-1].replace("월","")
```

```python
def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year

y = parse_year(date)
print(type(y))
y
```

```python
def parse_month(date):
    month = date.split("년")[-1].replace("월","")
    month = int(month)
    return month
m = parse_month(date)
print(type(m))
m
```

```python
parse_month(date)
```

```python
# apply 함수를 통해 새로운 칼럼 생성
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt.head(1)
```

```python
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
```

```python
# df_last와 병합하기 위해서는 컬럼의 이름이 같아야 합니다.
# sample을 활용해서 데이터를 미리보기 합니다.
df_last.sample()
```

```python
df_last.columns.to_list()
```

```python
cols = ['지역명', '연도', '월','평당분양가격']
cols
```

```python
# df_last에는 전용면적이 있음 
# 이전 데이터는 없기 떄문에 "전체"만 사용하도록 함
# loc를 사용해서 전체에 해당하는 면적만 copy로 복사함
df_last_prepare = df_last.loc[df_last["전용면적"] == "전체", cols].copy()
df_last_prepare.head(1)
```

```python
df_first_prepare = df_first_melt[cols].copy()
df_first_prepare.head(1)
```

## concat으로 합치기

```python
# df_first_prepare 와 df_last_prepare를 합치기

df = pd.concat([df_first_prepare, df_last_prepare])
df.shape
```

```python
# 제대로 합쳐졌는지 미리보기
df.tail()
```

```python
# 연도 별로 데이터가 몇개씩 있는지 확인
df["연도"].value_counts(sort = False)
```
