```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
headers = {'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}

url = "http://www.espn.com/nba/salaries"

r= requests.get(url, headers=headers)
r.status_code
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
print(soup)
```

```python
player_info = soup.find_all('tr', class_ = ["oddrow", "evenrow"])
```

```python
player_info[0]
```

```python
len(player_info)
```

```python
name = []
position = []
team = []
salary = []
```

```python
for info in player_info:
    player = info.find_all('td')
    name.append(player[1].text)
    team.append(player[2].text)
    salary.append(player[3].text)

    
#player[0] = 인덱스 번호
#player[1] = 이름
#player[2] = 팀이름
#player[3] = 연봉
```

```python
name
```

```python
position = []
name2 = []
for i in range(40):
    a = name[i].split(", ")
    position.append(a[1])
    name2.append(a[0])
```

```python
df = pd.DataFrame(
    {"name" : name2,
     "position" : position,
     "team" : team,
     "salary" : salary}
)
df
```
