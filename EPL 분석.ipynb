```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
epl = pd.read_csv("epl.csv", sep = ",")
epl.head()
```

```python
# 결측치 값 0 으로 채우기
epl.fillna(0)
```

```python
epl["Position"].value_counts()
```

```python
# 공격수 모으기
epl_Forward = epl[(epl["Position"] == "Forward") & (epl["Appearances"] > 50)]
epl_Forward = epl_Forward.fillna(0)
```

```python
# 미드필더 모으기
epl_Midfielder = epl[(epl["Position"] == "Midfielder") & (epl["Appearances"] > 50)]
epl_Midfielder = epl_Midfielder.fillna(0)
```

```python
# 수비수 모으기
epl_Defender = epl[(epl["Position"] == "Defender") & (epl["Appearances"] > 50)]
epl_Defender = epl_Defender.fillna(0)
```

```python
# 골키퍼 모으기
epl_Goalkeeper = epl[(epl["Position"] == "Goalkeeper") & (epl["Appearances"] > 50)]
epl_Goalkeeper = epl_Goalkeeper.fillna(0)
```

```python
epl.columns
```

# --------------------------공격수 ------------------------------

```python
epl_Forward["경기당 유효슈팅 점수"] = epl_Forward["Shots on target"] / epl_Forward["Appearances"] * 0.5
epl_Forward["경기당 골 점수"] = epl_Forward["Goals"] / epl_Forward["Appearances"] * 1
epl_Forward["경기당 도움 점수"] = epl_Forward["Assists"] / epl_Forward["Appearances"] * 1
epl_Forward["경기당 경고 점수"] = epl_Forward["Yellow cards"] / epl_Forward["Appearances"] * -0.5
epl_Forward["경기당 퇴장 점수"] = epl_Forward["Red cards"] / epl_Forward["Appearances"] * -1
```

```python
epl_Forward["총점"] = epl_Forward["경기당 유효슈팅 점수"] + epl_Forward["경기당 골 점수"] + epl_Forward["경기당 도움 점수"] + epl_Forward["경기당 경고 점수"] + epl_Forward["경기당 퇴장 점수"] 
```

```python
cols1 = ["Name", "Club", "Appearances", "경기당 유효슈팅 점수", "경기당 골 점수", "경기당 도움 점수", 
        "경기당 경고 점수", "경기당 퇴장 점수", "총점"]
epl_Forward[cols1]
```

# ------------------------ 수비수 ------------------------------

```python
epl_Defender["경기당 태클 점수"] = epl_Defender["Tackles"] / epl_Defender["Appearances"] * 0.8
epl_Defender["경기당 슈팅저지 점수"] = epl_Defender["Blocked shots"] / epl_Defender["Appearances"] * 1
epl_Defender["경기당 가로채기 점수"] = epl_Defender["Interceptions"] / epl_Defender["Appearances"] * 1
epl_Defender["경기당 공을 걷어 낸 점수"] = epl_Defender["Clearances"] / epl_Defender["Appearances"] * 0.5
epl_Defender["경기당 볼 탈취 점수"] = epl_Defender["Recoveries"] / epl_Defender["Appearances"] * 1
epl_Defender["경기당 그라운드 볼 경합 승리 점수"] = epl_Defender["Duels won"] / epl_Defender["Appearances"] * 0.5
epl_Defender["경기당 그라운드 볼 경합 패배 점수"] = epl_Defender["Duels lost"] / epl_Defender["Appearances"] * -0.5
epl_Defender["경기당 실점으로 이어진 실책 점수"] = epl_Defender["Errors leading to goal"] / epl_Defender["Appearances"] * -5
```

```python
epl_Defender["총점"] = epl_Defender["경기당 태클 점수"] + epl_Defender["경기당 슈팅저지 점수"] + epl_Defender["경기당 가로채기 점수"] + epl_Defender["경기당 공을 걷어 낸 점수"] + epl_Defender["경기당 볼 탈취 점수"] + epl_Defender["경기당 그라운드 볼 경합 승리 점수"] + epl_Defender["경기당 그라운드 볼 경합 패배 점수"] + epl_Defender["경기당 실점으로 이어진 실책 점수"] 
```
